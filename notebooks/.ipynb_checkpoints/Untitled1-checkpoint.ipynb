```python
import pandas as pd
import numpy as np
import holoviews as hv
from math import pi

import plotly.offline as ply
import plotly.graph_objs as go
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
from bokeh.palettes import Viridis, Paired

init_notebook_mode(connected=True)
```

```python
data = pd.read_csv('/Users/michael.chestnut/Documents/DOI/data/cleaned_data.csv')
df = data.drop(columns='Unnamed: 0')
df.head(5)
```

```python
#Put this in a new aggregate function
donut=df.groupby(['bureau_name','portfolio_part','fea_service_area']).sum().reset_index()

#Don't need this
donut['trimmed_fea_service_area'] = donut['fea_service_area'].str.split('-',1).str[1]
donut
```

```python
def update_donut(bureau="Department of the Interior",fy="FY18"):
    filtered_df = donut[['bureau_name','trimmed_fea_service_area',fy]]
    filtered_df = filtered_df[filtered_df['bureau_name']==bureau]
    
    #Sort and return top 10 based on FY values
    top10 = filtered_df.nlargest(10,fy)
    return top10
#     data = dict(
#         values=top10[fy],
#         labels=top10['trimmed_fea_service_area'],
#         domain= dict(
#             x=[0,.48]
#         ),
#         name="DOI",
#         hoverinfo="label+value+name",
#         hole=.4,
#         type="pie"
#     )
#     layout = dict(
#         title=bureau + ' spending by FEA Service Area in ' + fy,
#         annotations=[dict(
#             font=dict(
#                 size=20
#             ),
#             showarrow=False,
#             text="DOI",
#             x=.39,
#             y=0.5
#         )],
#         margin = dict(l = 200, r = 0, b = 10, t = 30, pad = 0),
#     )
#     fig = go.Figure(data=[data],layout=layout)
#     ply.iplot(fig, filename='donut-chart')
```

```python
top10=update_donut()
```

```python
import textwrap
def wrap_legend_text(series):
    return ['<br>'.join(textwrap.wrap(string, width=26)) for string in series]

```
